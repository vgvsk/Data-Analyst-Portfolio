````python
#функция для подсчёта пропусков и их доли от общего кол-ва значений
def nan_values(df):
    count_missing = df.isna().sum()
    percent_missing = round(df.isna().sum() * 100 / len(df), 2)
    
    missing_value_df = pd.DataFrame({'num_missing': count_missing,
                                     'percent_missing': percent_missing})
    
    return missing_value_df
````

**Возможные причины отсутвия оценки игр со стороны критиков и игроков:**
- до начала 1999-го года не было сайтов, которые собирали бы оценки на видеоигры: первым таким сайтом-агрегатором стал GameRanking (1999-2019), затем в 2001 году появился самый популярный в индустрии игр сайт-агрегатор — Metacritic (судя 100-бальной системе, данные взяты именно с этого сайта-агрегатора);
- игры создавались для локальной аудитории и не получили широкой огласки: например, в таблице есть игры, продажи которых состоялись в конкретном регионе и не превысили 10 000 копий;
- игру ещё не успели оценить, так как она ещё не поступила в продажу и разработчики не высылали игру на рецензию игровым критикам.

**Возможные причины отсутствия рейтинга со стороны ESRB:**
- ESRB начала функционировать только с 1994-го года, и игры, которые были выпущены до этого года, могли не получить рейтинг ESRB;
- игры не распространялись или не были произведены на территории США (согласно [сайту ESRB](https://www.esrb.org/faqs/#are-all-games-required-to-have-a-rating), рейтинг ESRB является обязательным только для ритейлеров и компаний по производству игр в США).

**Возможные причины отсутствия года:**
- даты отсутствуют как для игр середины 2000-х, так и для игр 2010-х — возможно, что данные потерялись по каким-то техническим причинам.

### Проверка данных на аномалии и дубликаты

````python
#функция для проверки данных на наличие аномальных значений
def describe_cols(df):
    
    values = pd.DataFrame()
    
    for col in df.columns:
        values[col] = [df[col].unique(), df[col].nunique()]
        
    return values.transpose().sort_values(by=1, ascending=True).rename(
        {0: 'Значения', 1: 'Кол-во уникальных значений'}, axis=1)
````

## Предобработка данных

### Изменение имён колонок

````python
#смена регистра имён колонок 
games.columns = games.columns.str.lower()
games.columns
````

### Замена типа данных и пропусков, удаление дубликатов

````python
#замена типа данных в колонке `year_of_release` на тип datetime
games['year_of_release'] = pd.to_datetime(games['year_of_release'], format='%Y', errors='coerce')
games['year_of_release'] = games['year_of_release'].dt.year
games.head(5)
````

````python
#замена NaN на 0 в столбцах `critic_score`, `user_score`
games['critic_score'] = games['critic_score'].fillna(0)
games['user_score'] = games['user_score'].fillna(0)
````

````python
#замена пропусков на значение unknown
games['rating'] = games['rating'].fillna("unknown")
````

````python
#замена значения "tbd" на 0
games['user_score'] = games['user_score'].replace('tbd', -1)
````

````python
#замена типа данных в колонке `user_score' на int
games['user_score'] = games['user_score'].astype(float)
````

4. В 2001 году в конукуренцию с Sony и Nintendo вступает **Microsoft** и выпускает собственную консоль Xbox.
5. В борьбе с Sony и Microsoft Nintendo фокусируется на **казуальном гейминге** — гейминге, который доступен для широкой аудитории, и создаёт Nintendo DS (2004) и Nintendo Wii (2006).
6. В 2005-2009 происходит **«инди-революция»**: появляются бесплатные игровые движки вроде Unity, а также платформы для распространения видеоигр вроде Steam. Аналогичный Steam'у сервис развивается у Sony и Apple.

#### Влияние рейтинга ESRB на продажи по регионам

````python
#создание функции, возвращающей по отметке рейтинга ESRB начальный возраст
#RP — рейтинг ещё не назначен, поэтому unknown 
def esrb_category (esrb):
    if esrb == 'EC':
        return 3
    elif esrb == 'E':
        return 0
    elif esrb == 'E10+':
        return 10
    elif esrb == 'T':
        return 13
    elif esrb == 'M':
        return 17
    elif esrb == 'AO':
        return 18
    elif esrb == 'unknown':
        return -1
    else:
        return "unknown"
````

````python
#создание колонки с начальными возрастами, соответствующими рейтингу ESRB
games_actual['esrb_category'] = games_actual.apply(lambda x: esrb_category(x['rating']), axis = 1)
````

````python
#создание выборки без строк, где рейтинг не указан
games_actual_esrb = games_actual[games_actual['esrb_category'] != 'unknown']
#изменение типа данных на целочисленный для значений рейтинга
games_actual_esrb['esrb_category'] = games_actual_esrb['esrb_category'].astype(int)
````

````python
#анализ суммарных продаж в Японии по рейтингу ESRB
games_actual_esrb.groupby('esrb_category')['jp_sales'].sum().sort_values(ascending = False) / games_actual_esrb['jp_sales'].sum() * 100
````

````python
#анализ суммарных продаж в Северной Америке по рейтингу ESRB
games_actual_esrb.groupby('esrb_category')['na_sales'].sum().sort_values(ascending = False) / games_actual_esrb['na_sales'].sum() * 100
````

````python
#анализ суммарных продаж в ЕС по рейтингу ESRB
games_actual_esrb.groupby('esrb_category')['eu_sales'].sum().sort_values(ascending = False) / games_actual_esrb['eu_sales'].sum() * 100
````

## Статистический анализ данных

### Проверка гипотезы о равенстве пользовательских рейтингов игр платформ XboxOne и PC

#### Проверка размера выборок

````python
#проверка количество значений в выборке по XOne
games_actual[games_actual['platform'] =='XOne'].shape
````

````python
#проверка количество значений в выборке по PC
games_actual[games_actual['platform'] =='PC'].shape
````

Можно делать **t-тест для двух генеральных совокупностей**:

1. Они независимы: средние не берутся из одной и той же генеральной совокупности до и после её изменения.
2. Выборочные средние, которые проверяются, нормально распределены. Согласно ЦПТ, если размер выборки составляет хотя бы несколько десятков значений, выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности. Выборочные средние берутся из выборок, где n1 = 247, n2 = 389.

#### Формулирование и проверка статистических гипотез

```
H_0: Пользовательский рейтинг игр на платформе XboxOne = пользовательский рейтинг игр на платформе PC`
H_a: Пользовательский рейтинг игр на платформе XboxOne ≠ пользовательский рейтинг игр на платформе PC`
alpha = 0.05
```

````python
#проверка гипотезы, что средние пользовательские рейтинги игр, сделанных на платформах Xbox One и PC одинаковые
results = st.ttest_ind(games_actual[games_actual['platform'] =='XOne']['user_score'], 
                      games_actual[games_actual['platform'] =='PC']['user_score'],
                      equal_var=False)
alpha = .05

print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Принимаем нулевую гипотезу')
````

Можно говорить, что гипотезу о равенстве двух выборок не удалось опровергнуть. Значит пользовательский рейтинг игр с большой вероятностью не зависит от платформы (PC или XboxOne).
